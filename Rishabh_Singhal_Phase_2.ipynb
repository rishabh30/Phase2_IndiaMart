```python
#np.quantile returns the value which divide the x into the respective ratios 
#IQR = Interquartile range 
#function to calculate appropriate unit wise price range which returns the upper, lower value and three most common price.
def IQR(x):
    Q1 = np.quantile(x, 0.375)
    Q3 = np.quantile(x, 0.875)
    Q4 = np.quantile(x, 0.125)
    Q5 = np.quantile(x, 0.375)
    IQR = Q3 - Q1
    IQR2 = Q5 - Q4
    lower = Q1 - 1.95*IQR2
    higher = Q3 + 1.8*IQR
    lower  = max(lower,x[0])
    higher = min(higher,x[len(x)-1])
    i=0
    j=len(x)-1
    while(lower>x[i]):
        i=i+1
    while(higher<x[j]):
        j=j-1
    k = int((j-i)/3)
    h = [x[i] for i in range(i,i+k)]
    i=i+k
    if(len(h)>0):
        v1 = statistics.mean(h)
    else:
        v1 = x[0]
    h = [x[i] for i in range(i,i+k)]
    if(len(h)>0):
        v2 = statistics.mean(h)
    else:
        v2 = v1
    i=i+k
    h = [x[i] for i in range(i,i+k)]
    if(len(h)>0):
        v3 = statistics.mean(h)
    else:
        v3 = v2
    return(lower, higher , v1, v2, v3)
```

```python
#some more modifications
def Basic_modifications(s):
#    for i in range(len(s)):
    s = s.replace("1 ","")
    s = s.replace("(S)","")
    s = s.replace("1","")
    s = s.replace("PER ","")
    s = s.replace("_","")
    s = s.replace(" ","")
    s = s.replace("LITRE","LITER")
    s = s.replace("ONWARD","")
    s = s.replace("MITRE","MITER")
    
    if(s[len(s)-1]=='S'):
        a = s
        s=a[:-1]
    return(s)
```

```python
#advance manual modifications
def Advance_modifications(s):
    s = s.replace("SQ.","SQUARE")
    s = s.replace("FT.","FEET")
    if(s=='MTR'):
        s = s.replace("MTR","METER")
    if(s=='KG'):
        s = s.replace("KG","KILOGRAM")
    if(s=='LTR'):
        s = s.replace("LTR","LITER")
    if(s=='NO'):
        s = s.replace("NO","NUMBER") 
    if(s=='SQUAREFEE'):
        s = s.replace("SQUAREFEE","SQUAREFEET")
    if(s=='PC'):
        s = s.replace("PC","PIECE")
    if(s=='PEICE'):
        s = s.replace("PEICE","PIECE")
    
    return(s)
```
